# 1. Basic plotting

## Matplotlib offers several ways to plot data. Here we will explore the common methods

### matplotlib.pyplot.plot()
#### Plot y versus x with a specified format
#### See https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
matplotlib.pyplot.plot(*args, scalex=True, scaley=True, data=None, **kwargs)

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y = 2*x+5 # Create the 'y' axis - y = 2x + 5 (linear)
plt.plot(x, y)
```

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y = np.sin(x)  # Create the 'y' axis - y = sin(x)
plt.plot(x, y, '.g') # plot green dots
```

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y = np.sin(x)  # Create the 'y' axis - y = sin(x)
plt.plot(x, y, 'bo') # plot blue circles
```

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y = np.sin(x)  # Create the 'y' axis - y = sin(x)
plt.plot(x, y, 'r+') # plot red pluses
```

### matplotlib.axes.Axes.scatter()
#### Plot y versus x scattered with a specified format and varying marker size and/or color
#### See https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html
Axes.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, *, edgecolors=None, plotnonfinite=False, data=None, **kwargs)

```python
x, y = np.random.rand(2, 30) # Create the x, y data - random between [0,1)
c = np.random.randint(1, 4, size=30) # Create an array of integers, random between [1,3), each will be mapped to a color
s = np.random.randint(1, 100, size=30) # Create an array of integers, random between [1,100), each will be mapped to a size
fig, ax = plt.subplots() # Create the figure and axes objects
ax.scatter(x, y, s=s, c=c) # Create the scatter
plt.show() # Plot the scatter
```

```python
fig, ax = plt.subplots() # Create the figure and axes objects
for color in ['tab:blue', 'tab:red', 'tab:green']: # for each color in the color array - blue, red, green, do:
    n = 100
    x, y = np.random.rand(2, n)  # Create the x, y data - 100 points in random between [0,1)
    scale = 200.0 * np.random.rand(n) # Create 100 random size values between 100*[0,1)
    ax.scatter(x, y, c=color, s=scale, label=color) # Create the scatter
plt.show() # Plot the scatter
```

### matplotlib.axes.Axes.bar()
#### Plot a bar of y versus x
#### See https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html
Axes.bar(x, height, width=0.8, bottom=None, *, align='center', data=None, **kwargs)

```python
x = 0.5 + np.arange(8) # Create an array for the 'x' axis between 0.5 to 8.5, +0.5 since the bar is +-0.5 wide to each side
y = np.random.uniform(2, 7, len(x)) # Create a uniformally distributed random array for the 'y' axis values
fig, ax = plt.subplots()
ax.bar(x, y, width=1, edgecolor="white")
plt.show()
```

```python
x = 0.5 + np.arange(8) # Create an array for the 'x' axis between 0.5 to 8.5, +0.5 since the bar is +-0.5 wide to each side
y = np.random.uniform(0, 7, len(x)) # Create a uniformally distributed random array for the 'y' axis values between [0,7) 
fig, ax = plt.subplots()
ax.bar(x, y, width=1, edgecolor="white") # Plot the bar diagram
ax.set(xlim=(0, 8), xticks=np.arange(1, 8), ylim=(0, 8), yticks=np.arange(1, 8)) # Arrange the axis limits and ticks
plt.show()
```

### matplotlib.axes.Axes.stem()
#### Create a stem plot (deltas, discrete)
#### See https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.stem.html
Axes.stem(*args, linefmt=None, markerfmt=None, basefmt=None, bottom=0, label=None, use_line_collection=True, orientation='vertical', data=None)

```python
x = np.arange(8)
y = np.random.uniform(2, 7, len(x))
fig, ax = plt.subplots()
ax.stem(x, y, 'r')
ax.set(xlim=(-1, 8), xticks=np.arange(-1, 9), ylim=(0, 8), yticks=np.arange(1, 8)) # set the 'x' limits to [-1,8] to show the 0
plt.show()
```

```python
x = np.arange(8)
y = np.random.uniform(2, 7, len(x))
fig, ax = plt.subplots()
ax.stem(x, y, '--g') # dashed green lines
ax.set(xlim=(-1, 8), xticks=np.arange(-1, 9), ylim=(0, 8), yticks=np.arange(1, 8)) # set the 'x' limits to [-1,8] to show the 0
plt.show()
```

### matplotlib.axes.Axes.step()
#### Create a step plot
#### See https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.step.html
Axes.step(x, y, *args, where='pre', data=None, **kwargs)

```python
x = np.arange(8)
y = np.random.uniform(2, 7, len(x))
fig, ax = plt.subplots()
ax.step(x, y)
ax.set(xlim=(0, 7), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
plt.show()
```

### Multiplte plots in the same figure

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y1 = x
y2 = x**2
y3 = x**3
plt.plot(x, y1, 'r') # plot y = x in red
plt.plot(x, y2, 'g') # plot y = x^2 in green
plt.plot(x, y3, 'b') # plot y = x^3 in blue
```

### matplotlib.pyplot.subplots()
#### Create multiple subplots
#### See https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html
matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y1 = x
y2 = x**2
y3 = x**3
plt.figure(figsize=(12, 3)) # Create a figure with set size
plt.subplot(131) # the 1st in 1 row, 3 columns grid of subplot
plt.plot(x, y1, 'r') # plot y = x in red
plt.subplot(132) # the 2nd in 1 row, 3 columns grid of subplot
plt.plot(x, y2, 'g') # plot y = x^2 in green
plt.subplot(133) # the 3rd in 1 row, 3 columns grid of subplot
plt.plot(x, y3, 'b') # plot y = x^3 in blue
```

# 2. Basic plot properties/attributes

### matplotlib.axes.Axes.legend()
#### Create and design the legend
#### See https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.legend.html
Axes.legend(*args, **kwargs)

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y1 = x
y2 = x**2
fig, ax = plt.subplots()
ax.plot(x, y1, 'r', label='y = x') # plot y = x in red
ax.plot(x, y2, 'g', label='y = x^2') # plot y = x^2 in green
ax.legend() # The labels are located at the plots
plt.show()
```

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y1 = x
y2 = x**2
fig, ax = plt.subplots()
ax.plot(x, y1, 'b') # plot y = x in red
ax.plot(x, y2, 'r') # plot y = x^2 in green
ax.legend(['y = x', 'y = x^2']) # The labels are located at ax.legend
plt.show()
```

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y1 = x
y2 = x**2
fig, ax = plt.subplots()
ax.plot(x, y1, 'b') # plot y = x in red
ax.plot(x, y2, 'r') # plot y = x^2 in green
ax.legend(['y = x', 'y = x^2'], loc='upper right') # Locate the legend at the upper right corner
plt.show()
```

```python
x = np.linspace(0, 10, 100) # Create a dataset of 100 points in the range [0,10) for the 'x' axis
y1 = x
y2 = x**2
fig, ax = plt.subplots()
ax.plot(x, y1, 'b') # plot y = x in red
ax.plot(x, y2, 'r') # plot y = x^2 in green
ax.legend(['y = x', 'y = x^2'], loc='center') # Locate the legend at the center
plt.show()
```
